```python
import pandas as pd
from fuzzywuzzy import fuzz, process
import numpy as np
```

```python
url = 'example.xlsx'
```

```python
df1 = pd.read_excel(url, sheet_name = 0)
df1
```

```python
df2 = pd.read_excel(url, sheet_name = 1)
df2
```

```python
df2["Best Match"] = df2["Sample Name/Location"].map(lambda x: process.extractOne(x, df1["Sample Name/Location"], scorer=fuzz.partial_ratio)[0])
df2
```

```python
def match_or_nan(x):
    match = process.extractOne(x, df1['Sample Name/Location'], scorer=fuzz.partial_ratio)
    if match[1] < 80:
        return np.nan
    else:
        return match[0]
```

```python
df2['Best Match'] = df2['Sample Name/Location'].apply(match_or_nan)
```

```python
df2 = df2.join(df1.set_index('Sample Name/Location'), on='Best Match', lsuffix='_left', rsuffix='_right')
df2
```
